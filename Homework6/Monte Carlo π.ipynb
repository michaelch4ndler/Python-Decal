```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimate_pi(N):
    points = np.random.rand(N, 2)
    inside_circle = points[:, 0]**2 + points[:, 1]**2 <= 1
    pi_estimate = 4 * np.sum(inside_circle) / N
    return pi_estimate

N_values = [int(1e3), int(1e4), int(1e5), int(1e6)]

for N in N_values:
    pi_estimate = estimate_pi(N)
    print(f"N = {N}: π ≈ {pi_estimate}")

N = int(1e4)
points = np.random.rand(N, 2)
inside_circle = points[:, 0]**2 + points[:, 1]**2 <= 1

plt.figure(figsize=(6, 6))  
plt.scatter(points[:, 0], points[:, 1], c=inside_circle, cmap='viridis')
plt.axis('equal')

pi_estimate = estimate_pi(N)
plt.text(0.05, 0.9, f'π ≈ {pi_estimate}', transform=plt.gca().transAxes, fontsize=12)

plt.title(f'Monte Carlo Estimation of π (N = {N})')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
