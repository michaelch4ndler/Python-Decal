```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = np.random.randint(0, 101, 40)
data2 = np.random.randint(0, 101, 40) 
data3 = np.random.randint(0, 101, 40)
fig, axs = plt.subplots(2, 1)
axs[0].plot(data1, color='orange', linewidth=10)
axs[0].plot(data2, color='red', linestyle='--')
axs[1].scatter(range(40), data3, marker='D', color='magenta')
axs[0].set_title('Random Data Plots')
axs[0].set_ylabel('Values')
axs[1].set_ylabel('Values')
axs[1].set_xlabel('Index')
plt.tight_layout()
plt.show()
```
