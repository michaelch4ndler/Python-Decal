```python
import seaborn as sns
import numpy as np
```

```python
planets_df = sns.load_dataset('planets')
```

```python
print(planets_df)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='orbital_period', y='mass', hue='method', data=planets_df)
plt.xscale('log')
plt.yscale('log')
plt.title('Orbital Period vs. Mass of Exoplanets')
plt.xlabel('Orbital Period (days)')
plt.ylabel('Mass (Jupiter mass)')
plt.legend(title='Discovery Method')
plt.show()
```

```python
cleaned_planets_df = planets_df.dropna(subset=['year'])
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='year', hue='method', data=cleaned_planets_df)
plt.title('Number of Exoplanets Discovered by Year')
plt.xlabel('Year')
plt.ylabel('Number of Exoplanets')
plt.legend(title='Discovery Method')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
